```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
```

```python
df = pd.read_csv("../traces/benchmarks/20230222_benchmarks_except_flowchart.csv")
```

```python
df['benchmark'].unique()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

axs[0].hist(df[df['benchmark'] == 'pow_2_3']['time'], bins = 20, range=(2e5, 6e5))
axs[1].hist(df[df['benchmark'] == 'pow_2_3_compile']['time'],  bins = 20, range=(2e5, 6e5))
axs[2].hist(df[df['benchmark'] == 'pow_2_3_staged']['time'], bins=40, range=(0, 25000))
plt.show()
```
